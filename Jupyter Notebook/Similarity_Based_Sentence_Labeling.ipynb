```python
# df에서 sentence 컬럼 추출 후 리스트로 변경
import numpy as np
lines = np.array(df.sentence).tolist()
```

```python
# word2vec 학습을 위해 2차원 리스트에 한 자리 단어와 숫자를 제거해 명사를 추출 후 데이터셋으로 저장
dataset = []
for i in range(len(lines)):
  dataset.append(komoran.nouns(lines[i]))
dataset = [[y for y in x if not len(y)==1] for x in dataset] # 1자 단어 제거
dataset = [[y for y in x if not y.isdigit()] for x in dataset] # 숫자 제거
dataset[:10]
```

```python
# 모델 구축
# Skip-Gram 적용: 중심 단어로 주변 단어 예측
# 중심 단어로부터 좌우 5개 단어까지 학습에 적용
# 전체 문서에서 최소 1회 이상 출현 단어로 학습 진행
model = Word2Vec(dataset,sg=1,window=5,min_count=1)
model.init_sims(replace=True)
```

```python
# 문장에서 명사를 추출한 리스트가 주어졌을 때, 리스트 내의 명사들을 벡터화
# 각 명사들의 벡터값 평균을 앞서 정의한 어려운 단어 대표 벡터와 비교해 유사도를 반환하는 함수
from sklearn.metrics.pairwise import cosine_similarity

# x는 한 문장을 명사로 변환한 1차원 리스트
def calculate_similarity(x):
  global similarity
  word_vectors = [loaded_model.wv[word] for word in x if word in loaded_model.wv]
  if word_vectors:
    sentence_mean_vector = np.mean(word_vectors, axis=0)
    similarity = cosine_similarity([sentence_mean_vector], [difficult_words_mean_vector])
  return similarity[0][0]
```

```python
# 각 카테고리에 해당하는 명사 추출 2차원 리스트가 주어졌을 때 인덱스, 유사도 순의 값이 저장된 2차원 리스트를 반환하는 함수
def category_similarity(x):
  li=[]
  for i in range(len(x)):
    sim = calculate_similarity(x[i])
    a=[i,sim]
    li.append(a)
  return li
```

```python
# 카테고리 별 [인덱스,유사도] 리스트를 요소로 가지는 2차원 리스트 생성
IT_science_sim = category_similarity(IT_science_nouns)
economy_sim = category_similarity(economy_nouns)
international_sim = category_similarity(international_nouns)
politics_sim = category_similarity(politics_nouns)
```

```python
# 문장에서 명사를 추출한 리스트가 주어졌을 때, 리스트 내의 명사들을 벡터화
# 각 명사들의 벡터값 평균을 앞서 정의한 어려운 단어 대표 벡터와 비교해 유사도를 반환하는 함수
from sklearn.metrics.pairwise import cosine_similarity

# x는 한 문장을 명사로 변환한 1차원 리스트
def calculate_similarity(x):
  global similarity
  word_vectors = [loaded_model.wv[word] for word in x if word in loaded_model.wv]
  if word_vectors:
    sentence_mean_vector = np.mean(word_vectors, axis=0)
    similarity = cosine_similarity([sentence_mean_vector], [easy_words_mean_vector])
  return similarity[0][0]
```

```python
# 각 카테고리에 해당하는 명사 추출 2차원 리스트가 주어졌을 때 인덱스, 유사도 순의 값이 저장된 2차원 리스트를 반환하는 함수
def category_similarity(x):
  li=[]
  for i in range(len(x)):
    sim = calculate_similarity(x[i])
    a=[i,sim]
    li.append(a)
  return li
```

```python
# 카테고리 별 [인덱스,유사도] 리스트를 요소로 가지는 2차원 리스트 생성
IT_science_sim = category_similarity(IT_science_nouns)
economy_sim = category_similarity(economy_nouns)
international_sim = category_similarity(international_nouns)
politics_sim = category_similarity(politics_nouns)
```
